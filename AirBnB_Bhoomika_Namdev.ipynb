## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
neighborhood_avg_price = Airbnb_df.groupby('neighbourhood')['price'].mean().reset_index()
neighborhood_avg_price
````

````python
# Define the minimum number of reviews and maximum price
min_reviews = 50
max_price = 200

# Filter the data based on the criteria
filtered_data = Airbnb_df[(Airbnb_df['number_of_reviews'] >= min_reviews) & (Airbnb_df['price'] <= max_price)]

# Displaying the filtered data
print(filtered_data)
````

### What all manipulations have you done and insights you found?

In our first manipulation we have done as mention : groupby('neighbourhood') groups the data by the 'neighbourhood' column, and ['price'].mean() calculates the mean price for each group. The reset_index() function resets the index of the resulting DataFrame for easier readability.

And in another manipulation we have filtered_data that will contain only the rows where the number of reviews is greater than or equal to min_reviews and the price is less than or equal to max_price.

#### Chart - 11

````python
# Chart - 11 visualization code
# Create a heatmap
corr = Airbnb_df[Airbnb_df.describe().columns].corr()
# Create a beautiful heatmap
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.2)  # Increase font size for readability

# Define custom color palette with more contrast
cmap = sns.diverging_palette(220, 20, as_cmap=True)

# Draw the heatmap with improved annotation and aesthetics
sns.heatmap(corr, cmap=cmap, annot=False, fmt=".2f", annot_kws={"size": 12, "weight": "bold"},
            linewidths=0.5, cbar_kws={"shrink": 0.8, "label": "Correlation Coefficient"},
            square=True, linecolor='white', vmin=-1, vmax=1)

# Add title
plt.title('Correlation Heatmap', fontsize=18)

# Rotate y-axis labels for better readability
plt.yticks(rotation=0)

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
````

##### 1. Why did you pick the specific chart?

I used a heatmap to visualize the correlation matrix of numerical columns in the Airbnb dataset.

##### 2. What is/are the insight(s) found from the chart?

The heatmap helps you identify which pairs of columns are strongly correlated, which can indicate potential relationships between variables. For example, if the 'price' column is strongly correlated with the 'availability_365' column, it suggests that prices may vary based on the availability of listings.
